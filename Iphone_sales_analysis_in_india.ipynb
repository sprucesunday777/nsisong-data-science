```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Let read our data in form of CSV file

```python
df=pd.read_csv('/content/apple_products.csv')
df.head()
```

```python
plt.figure (figsize=(20,5))
df.sort_values(['Star Rating'][:10], axis=0, ascending=False, inplace=True)
plt.xticks(rotation='vertical')
sns.barplot(x=df['Product Name'][:10],y='Star Rating',data=df)
plt.title('TOP 10 APPLE PRODUCT Name WITH THE HIGHEST STAR RATING',fontsize=40)
plt.show()
```


According to the above data, below are the top 5 most liked iPhones in India:
APPLE iPhone 11 Pro Max (Midnight Green, 64 GB) APPLE iPhone 11 Pro Max (Space Grey, 64 GB) APPLE iPhone 11 Pro Max (Midnight Green, 256 GB) APPLE iPhone 11 Pro Max (Gold, 64 GB) APPLE iPhone 11 Pro Max (Gold, 256 GB)

```python
plt.figure (figsize=(20,5))
df.sort_values(['Sale Price'], axis=0, ascending=False, inplace=True)
plt.xticks(rotation='vertical')
sns.barplot(x=df['Product Name'][:10],y='Sale Price',data=df)
plt.title('TOP APPLE PRODUCT NAME WITH THE HIGHEST SALE PRICE',fontsize=40)
plt.show()
```

From the result above the top 10 apple product with the highest price are APPLE IPHONE 12 PRO(PACIFIC,512GB)

```python
df.columns
```

```python
plt.figure (figsize=(20,5))
df.sort_values([ 'Number Of Reviews'], axis=0, ascending=False, inplace=True)
plt.xticks(rotation='vertical')
sns.barplot(x=df['Product Name'][:10],y='Number Of Reviews',data=df)
plt.title('TOP APPLE PRODUCT NAME WITH THE HIGHEST Number Of Reviews',fontsize=40)
plt.show()
```

According to the above bar graph, APPLE iPhone SE(BLACK,64 GB) has the most review on Flipkart.

```python
plt.figure (figsize=(10,5))
plt.xticks(rotation='vertical')
sns.regplot(y='Sale Price',x='Number Of Ratings',data=df)
plt.title('RELATIONSHIP BETWEEN SALE PRICE AND NUMBER OF RATING',fontsize=40)
plt.show()
```

From the result above,There is a negative linear relationship between the sale price of iPhones and the number of ratings. It means iPhones with lower sale prices are sold more in India.

```python
plt.figure (figsize=(10,5))
plt.xticks(rotation='vertical')
sns.regplot(y='Discount Percentage',x='Number Of Ratings',data=df)
plt.title('RELATIONSHIP BETWEEN Discount Percentage AND NUMBER OF RATING',fontsize=40)
plt.show()
```

There is a linear relationship between the discount percentage on iPhones on Flipkart and the number of ratings. It means iPhones with high discounts are sold more in India
