Web scraping from cars.com website

The Aim of this project was to scrap data from cars.com website, not all the data but only on merecedes-benz and we needed only the name of the car,rating,rating count,price and dealer name of merecedes-benz car

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D=&list_price_max=&maximum_distance=20&zip=60606'
```

```python
response=requests.get(url)
```

```python
response.status_code
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
results=soup.find_all('div',{'class','vehicle-card'})
```

Looping everything

```python
name=[]
mileage=[]
dealer_name=[]
rating=[]
rating_count=[]
price=[]
for result in results:
        name.append(result.find('h2').get_text())
        dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
        rating.append(result.find('span',{'class':'sds-rating__count'}).get_text ())
        mileage.append(result.find('div',{'class':'mileage'}).get_text ())
        rating_count.append(result.find('span',{'class':'sds-rating__link'}).get_text ())
        price.append(result.find('span',{'class':'primary-price'}).get_text())
```

```python
car_dealer=pd.DataFrame({'Name':name,'Dealer_name':dealer_name,'Rating':rating,'Mileage':mileage,'Rating_count':rating_count,'Price':price})
```
